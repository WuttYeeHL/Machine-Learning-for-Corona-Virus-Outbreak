```python
# importing the stuff we need
import pandas as pd
import numpy as np
df = pd.read_csv('data\diagnosis-of-covid-19-and-its-clinical-spectrum.csv')
df
```

```python
print(df.patient_id.nunique()) #checking if patient ID is really 5644
df = df.iloc[:,1:] #dropping the patient id because it doenst help in differentiation
df = df.replace(to_replace = ["negative","positive"], value =[0,1]) #changing string to int
df
```

```python
row_count = {} #setting up the dictionary of column name vs data count

#a loop to count all non null data rows on per column basis
for i in np.array(df.columns):
#     print('Column "',i,'" have', df[i].count() ,'values')
    row_count.update({i:df[i].count()})

#previewing the column vs non-null row count
for x,y in row_count.items():
    print(y,'\t',x) 
```

```python
# sorting the columns based on number of rows

row_count_values = [] #instantiating a list for sorting

for x,y in row_count.items():
    row_count_values.append(y)
    
#sorting the unique number of rows in the array in descdening order    
sorted_row_count = np.sort(np.unique(row_count_values))[::-1]

#previwing the array
print(sorted_row_count)
```

```python
#defining a function which yields the most number of columns without losing rows

def data_filter(upperlimit,lowerlimit):
    columncount = 0
    selected_column = ['sars_cov_2_exam_result']
    for column_name, count in row_count.items():
        if count >= lowerlimit and count <= upperlimit:
            if column_name != 'sars_cov_2_exam_result':
                selected_column.append(column_name)  
    for i in selected_column:
        if i != 'sars_cov_2_exam_result' and df.loc[(df[i].notnull()),i].unique().size > 1:
            columncount += 1
    
    newdf = df.loc[:,selected_column]
    newdf.dropna(subset=selected_column,inplace=True)
    row,column = newdf.shape
    if row >= lowerlimit*0.8 and row<=upperlimit:
        return 0
    else:
        return -1
```

```python
lowerindex = 2
upperindex = 1
combo = []
while upperindex < len(sorted_row_count) and lowerindex <len(sorted_row_count):
    if data_filter(sorted_row_count[upperindex],sorted_row_count[lowerindex]) == 0:
        lowerindex +=1
    else:      
        combo.append([sorted_row_count[upperindex],sorted_row_count[lowerindex-1]])
        upperindex = lowerindex
        lowerindex +=1
print(combo)
```

```python
def shape_finder(upperlimit,lowerlimit):
    columncount = 0
    selected_column = ['sars_cov_2_exam_result']
    for column_name, count in row_count.items():
        if count >= lowerlimit and count <= upperlimit:
            if column_name != 'sars_cov_2_exam_result':
                selected_column.append(column_name)  

    for i in selected_column:
        if i != 'sars_cov_2_exam_result' and df.loc[(df[i].notnull()),i].unique().size > 1:
            columncount += 1
    
    newdf = df.loc[:,selected_column]

    newdf.dropna(subset=selected_column,inplace=True)
    row,column = newdf.shape
    print('Number of positives:', newdf.loc[newdf['sars_cov_2_exam_result']==1,'sars_cov_2_exam_result'].count())
    print('Number of negatives:', newdf.loc[newdf['sars_cov_2_exam_result']==0,'sars_cov_2_exam_result'].count())
    print('Positive to Negative percentage:', newdf.loc[newdf['sars_cov_2_exam_result']==1,'sars_cov_2_exam_result'].count()/newdf.loc[newdf['sars_cov_2_exam_result']==0,'sars_cov_2_exam_result'].count()*100)
    print(row,'rows x ',column,'columns','\n')
    return
```

```python
def data_maker(upperlimit,lowerlimit):
    columncount = 0
    selected_column = ['sars_cov_2_exam_result']
    for column_name, count in row_count.items():
        if count >= lowerlimit and count <= upperlimit:
            if column_name != 'sars_cov_2_exam_result':
                selected_column.append(column_name)  

    for i in selected_column:
        if i != 'sars_cov_2_exam_result' and df.loc[(df[i].notnull()),i].unique().size > 1:
            columncount += 1
    
    newdf = df.loc[:,selected_column]

    newdf.dropna(subset=selected_column,inplace=True)
    return newdf
```

```python
#as all the sub datasets generally represents a lesser number of positives
#we need to perform oversampling

def over_sampler(dfs):
    import random
    from imblearn.over_sampling import RandomOverSampler
    a = random.randint(0,100000)
    
    ros = RandomOverSampler(random_state=a)
    
#splitting the original dataset into positive and negative labels
    pos_x = dfs.iloc[:,1:].loc[dfs['sars_cov_2_exam_result']==1,:]
    neg_x = dfs.iloc[:,1:].loc[dfs['sars_cov_2_exam_result']==0,:]
    pos_y = dfs.iloc[:,[0]].loc[dfs['sars_cov_2_exam_result']==1,:]
    neg_y = dfs.iloc[:,[0]].loc[dfs['sars_cov_2_exam_result']==0,:]
    pos_y = pos_y.iloc[:,0] #need to change this to a panda series as system expect a 1d array 
    neg_y = neg_y.iloc[:,0] #need to change this to a panda series as system expect a 1d array
#     print('pos_y ',pos_y.shape)
#     print('neg_y ',neg_y.shape)

#splitting the subdivived dataset with the built in 25/75 ratio
    pos_x_train, pos_x_test, pos_y_train, pos_y_test = train_test_split(pos_x,pos_y,random_state=0)
    neg_x_train, neg_x_test, neg_y_train, neg_y_test = train_test_split(neg_x,neg_y,random_state=0)
#     print('pos_y_train ',pos_y_train.shape)
#     print('neg_y_train ',neg_y_train.shape)

#append all the evenly distributed labels and features back to a single array
    x_train = pos_x_train.append(neg_x_train)
    x_test = pos_x_test.append(neg_x_test)
    y_train = pos_y_train.append(neg_y_train)
    y_test = pos_y_test.append(neg_y_test)
#     print('y_train ',y_train.shape)

#removing the bias of the training dataset
    x_resampled, y_resampled = ros.fit_sample(x_train, y_train)
#     print('y_resampled ', y_resampled.shape)
#     print(collections.Counter(y_resampled))

#returning the over_sampled training and testing dataset
    return x_resampled,x_test, y_resampled,y_test
```

```python
#performing the pearson correlation matrix
def corrdf(df):
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    print('#With Correlation Matrix\n')
    
    corr_mat = df.corr()
    plt.figure(figsize=(15,6))
    
    sns.heatmap(data=corr_mat,annot=True,cmap='GnBu')
    plt.show()
    
    target = 'sars_cov_2_exam_result'
    features = corr_mat.index.values
    features = features[features != target]  
    
    removed = []
    
    for i in features:
        for j in features:
            if(j not in removed and i not in removed):
                if ((corr_mat[i].loc[[j]].values > 0.6) or (corr_mat[i].loc[[j]].values < -0.6)) and (corr_mat[i].loc[[j]].
                                                                                                      values != 1) :
                    removed.append(j)
    print('\nThe columns ',removed,'are removed\n')

    final_feature = [x for x in features if x not in removed]
    final_feature.insert(0,target)
    
    return df.loc[:,final_feature]
    
    
```

```python
#performing PCA

def pcadf(df):
    
    from sklearn.decomposition import PCA
    
    print('#With PCA\n')
    
    best_n = {}
    for i in range(1,len(df.columns)):
        pca = PCA(n_components=i)
        pc = pca.fit_transform(df.iloc[:,1:])
        ratio_sum = pca.explained_variance_ratio_.sum()
        if ratio_sum > 0.85: #we only consider sum above 0.85
            best_n.update({i:ratio_sum})
    
    best_ratio_sum = min(best_n.values())
    
#     print(best_ratio_sum)
    
    for n,s in best_n.items():
#         print(n , 'components yield', s,'variance ratio sum') #can print this line if you want the details
        if s == best_ratio_sum:
            bestN=n  

    pca = PCA(n_components=bestN)
    pc = pca.fit_transform(df.iloc[:,1:])
    
    print('The most optimal n is ',bestN,'\n')
    print('With variance ratio sum of ',pca.explained_variance_ratio_.sum(),'(minimum 0.85)\n')
    print('The variance ratios are\n', pca.explained_variance_ratio_,'\n')
    column_names = ['sars_cov_2_exam_result']
    for i in range(1,pc.shape[1]+1):
        column_names.append('PC'+str(i))
        
    pcadf = pd.DataFrame(np.append(np.reshape(df.iloc[:,[0]], (-1,1)), pc, axis=1))
    pcadf.columns = column_names
    return pcadf
```

# Part 3 model training and evaluation

```python
# parameter initialisers
def param_reset():
    LRacc = []
    LRst = []
    LRet = []
    KNNacc = []
    KNNst = []
    KNNet = []
    bestN = []
    DTacc = []
    DTst = []
    DTet = []
    NNacc = []
    NNst = []
    NNet = []
    return LRacc,LRst,LRet,KNNacc,KNNst,KNNet,bestN,DTacc,DTst,DTet,NNacc,NNst,NNet
```

```python
def time_calculator(LRst,LRet,KNNst,KNNet,DTst,DTet,NNst,NNet):
    LRdiff=[]
    for s ,e in zip(LRst,LRet):
        LRdiff.append(e-s)

    KNNdiff=[]
    for s ,e in zip(KNNst,KNNet):
        KNNdiff.append(e-s)

    DTdiff=[]
    for s ,e in zip(DTst,DTet):
        DTdiff.append(e-s)

    NNdiff=[]
    for s ,e in zip(NNst,NNet):
        NNdiff.append(e-s)
        
    return LRdiff,KNNdiff,DTdiff,NNdiff
```

```python
#finding the best KNN model based on a input

def bestknn(x_train,y_train,x_test,y_test):
    import time
    bestacc = {}
    KNNst = []
    KNNet = []
    for i in range(1,int(x_train.shape[0]/2),2): 
        #presuming best n will never be more than half of training rows to prevent underfitting
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(x_train,y_train)
        KNNy_pred = knn.predict(x_test)
        KNNacc = accuracy_score(y_test,KNNy_pred)
        bestacc.update({i:KNNacc})
        
    bestnacc = max(bestacc.values())
    
    for n,acc in bestacc.items():
#         print(n,acc) #only enable for single run else it still stall the loop
        if acc == bestnacc:
            bestn = n
    
    knn = KNeighborsClassifier(n_neighbors=bestn)
    KNNst.append(time.time())
    knn.fit(x_train,y_train)
    KNNet.append(time.time())
    KNNy_pred = knn.predict(x_test)
    KNNacc = accuracy_score(y_test,KNNy_pred)
    
    return KNNy_pred,KNNacc,bestn,KNNst,KNNet
```

```python
#modularizing the neural network

def bestnn(x_train,y_train,x_test,y_test):
    NNst = []
    NNet = []
    y_train_ohe = [[1,0] if x == 0 else [0,1] for x in np.array(y_train)]
    y_train_ohe = np.array(y_train_ohe)
    
    y_test_ohe = [[1,0] if x == 0 else [0,1] for x in np.array(y_test)]
    y_test_ohe = np.array(y_test_ohe)
    
    model = Sequential()
    model.add(Dense(500, input_shape=(x_train.shape[1],),activation='sigmoid'))
    model.add(Dense(2,activation='softmax'))
    model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=["accuracy"])
    NNst.append(time.time())
    model.fit(x_train,y_train_ohe,batch_size=64,epochs=500,verbose=0)
    NNet.append(time.time())
    NNy_pred = model.predict(x_test)
    loss,NNindvaccuracy = model.evaluate(x_test,y_test_ohe,verbose=0)
    
    return y_test_ohe,NNst,NNet,NNy_pred,loss,NNindvaccuracy
    
    
```

```python
def result_printer(LRacc,LRdiff,LRy_pred,bestN,KNNacc,KNNdiff,KNNy_pred,
                   DTacc,DTdiff,DTy_pred,NNacc,NNdiff, NNy_pred):
    from sklearn.metrics import confusion_matrix
    
    list = []
    for i in np.arange(len(NNy_pred)):
        if NNy_pred[i,0] > NNy_pred[i,1]:
            list.append(0.0)
        else:
            list.append(1.0)
            
    print('After a 100 runs\n')
    
    print('#For Logistic Regresstion',
      '\nThe average accuracy score is ',np.mean(LRacc),
      '\nThe standard deviation for accuracy score is ',np.std(LRacc),
     '\nThe maximum accuracy is ',np.max(LRacc),
     '\nThe minimum accuracy is ',np.min(LRacc),
      '\nThe average time taken is ',np.mean(LRdiff),
          '\nThe confusion matrix for the last iteration is\n',
          confusion_matrix(y_test, LRy_pred, labels =[1, 0]),'\n')

    print('#For K Nearest Neighbours',
      '\nThe most common n is n = ',stats.mode(bestN)[0][0],
      '\nThe average accuracy score is ',np.mean(KNNacc),
      '\nThe standard deviation for accuracy score is ',np.std(KNNacc),
     '\nThe maximum accuracy is ',np.max(KNNacc),
     '\nThe minimum accuracy is ',np.min(KNNacc),
      '\nThe average time taken is ',np.mean(KNNdiff),'(excluding time to find the best n)',
         '\nThe confusion matrix for the last iteration is\n',
          confusion_matrix(y_test, KNNy_pred, labels =[1, 0]),'\n')

    print('#For Decision Tree',
      '\nThe average accuracy score is ',np.mean(DTacc),
      '\nThe standard deviation for accuracy score is ',np.std(DTacc),
     '\nThe maximum accuracy is ',np.max(DTacc),
     '\nThe minimum accuracy is ',np.min(DTacc),
      '\nThe average time taken is ',np.mean(DTdiff),
          '\nThe confusion matrix for the last iteration is\n',
          confusion_matrix(y_test, DTy_pred, labels =[1, 0]),'\n')

    print('#For Neural Network',
      '\nThe average accuracy score is ',np.mean(NNacc),
      '\nThe standard deviation for accuracy score is ',np.std(NNacc),
     '\nThe maximum accuracy is ',np.max(NNacc),
     '\nThe minimum accuracy is ',np.min(NNacc),
      '\nThe average time taken is ',np.mean(NNdiff),
          '\nThe confusion matrix for the last iteration is\n',
          confusion_matrix(y_test, list, labels =[1,0]),'\n')
    return
```

```python
import sklearn
import imblearn
import collections
import time
from scipy import stats
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from keras.layers.core import Dense, Activation
from keras.models import Sequential
```

```python
#the following test are for with correlation matrix and PCA (with correlation matrix first)

print('##Using both correlation matrix and PCA (with correlation matrix first)\n')
df=pcadf(corrdf(dfs[1]))

#reseting all parameters
LRacc,LRst,LRet,KNNacc,KNNst,KNNet,bestN,DTacc,DTst,DTet,NNacc,NNst,NNet = param_reset()

#looping all 4 training models 100 times
for i in range(100):
    
    #preparing the common data
    x_train,x_test,y_train,y_test = over_sampler(df)
    
    
    #using Logistic Regression
    logReg = LogisticRegression(solver = 'lbfgs')
    LRst.append(time.time())
    logReg.fit(x_train,y_train)
    LRet.append(time.time())
    LRy_pred = logReg.predict(x_test)    
    LRacc.append(accuracy_score(y_test,LRy_pred))

    #using KNN    
    KNNy_pred,KNNsubacc,bestn,KNNst,KNNet = bestknn(x_train,y_train,x_test,y_test)
    bestN.append(bestn)
    KNNacc.append(KNNsubacc)
    
    #using DT
    dt = DecisionTreeClassifier(criterion = "gini", max_depth = 5)  
    DTst.append(time.time())
    dt.fit(x_train, y_train)
    DTet.append(time.time())
    DTy_pred = dt.predict(x_test)
    DTacc.append(accuracy_score(y_test,DTy_pred))
    
    #using NN       
    y_test_ohe,NNst,NNet,NNy_pred,loss,NNindvaccuracy = bestnn(x_train,y_train,x_test,y_test)
    NNacc.append(NNindvaccuracy)

# calculation the time taken for all 4 models
LRdiff,KNNdiff,DTdiff,NNdiff = time_calculator(LRst,LRet,KNNst,KNNet,DTst,DTet,NNst,NNet)

#printing the results for all 4 models
result_printer(LRacc,LRdiff,LRy_pred,bestN,KNNacc,KNNdiff,KNNy_pred,
                   DTacc,DTdiff,DTy_pred,NNacc,NNdiff, NNy_pred)
```
